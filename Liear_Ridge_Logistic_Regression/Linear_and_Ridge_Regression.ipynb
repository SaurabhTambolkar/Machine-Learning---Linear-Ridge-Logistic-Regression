```python
'''Importing Library'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''Converting SEX to number feature'''

df.sex = [1 if each == "female" else 0 for each in df.sex]
```

```python
'''Converting SMOKER to number feature'''

df.smoker = [1 if each == "yes" else 0 for each in df.smoker]
```

```python
'''Dropping Column Sex,children,region as the have a very correlation to charges'''

df = df.drop(["sex", "children", "region"], axis=1)
```

```python
'''Printing Head of the dataset'''
df.head
```

```python
'''Seprating Feature'''

x=df[['age','bmi','smoker']]
print(x)
```

```python
'''normailizing feature data '''

x_norm= x.copy()

for column in x_norm.columns:
    x_norm[column] = x_norm[column]  / x_norm[column].abs().max()
      
# view normalized data
display(x_norm)
```

```python
'''seprating target variable'''

y=df[['charges']]
print(y)
```

## Spliting the DATA in Test and Train Set

```python
'''Spliting the feature data'''
x_train = x_norm.sample(frac = 0.80)
x_test = x_norm.drop(x_train.index)

print(x_train.shape)
print(x_test.shape)
```

```python
'''Spliting the target data'''

y_train = y.sample(frac = 0.80)
y_test = y.drop(y_train.index)

print(y_train.shape)
print(y_test.shape)
```

**LINEAR REGRESSION**

**CalculateOLS Function**

```python
def CalculateOLS(x, y):
  r, c = x.shape
  w = np.dot(np.linalg.inv(np.dot(np.transpose(x),x)), np.dot(np.transpose(x),y))
  return w
```

**Predict_Y Function**

```python
def Predict_Y(w, x):
  return np.dot(x,w)
```

**Mean Square Loss Function**

```python
def Mean_Square_Loss(A,B):
  return np.square(A - B).mean()
```

**Calculate Weight Using OLS Function**

```python
w_l = CalculateOLS(x_train, y_train)
w_l
```

**Predict Output**

```python
y_pred_linear = Predict_Y(w_l,x_test)
y_pred_linear
```

**Calculate Mean Square Loss**

```python
loss = Mean_Square_Loss(y_test, y_pred_linear)
loss
```

**RIDGE REGRESSION**

**CalulateOLS function**

```python
def CalculateOLS(x, y, lam = 100):
  r, c = x.shape
  w = np.dot(np.linalg.inv(np.dot(np.transpose(x),x) + lam * np.identity(c)), np.dot(np.transpose(x),y))
  return w
```
