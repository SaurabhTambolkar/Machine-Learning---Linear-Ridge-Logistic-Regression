## Data Visualization-insurance.csv




```python
'''Importing Required Library'''

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import pie, axis, show
```

```python
''' Correlation heatmap '''

df.corr()
sns.heatmap(df.corr())
```

```python
''' Book Genre Pie-Ratio'''

genre1=df['Genre'].value_counts()
content=list(genre1.keys())

plt.pie(genre1,labels = content,autopct = '%.2f%%')
plt.title('Book Genre', fontsize = 20)
plt.show()
```

```python
''' BOOK COST VS NUMBER OF BOOK SOLED PLOT '''

price1=df['Price'].value_counts()
content=list(price1.keys())

fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.barh(content,price1)
plt.ylabel('Book Cost',fontsize=16); 
plt.xlabel('Number of Books Soled',fontsize=16)
plt.figure(figsize=(10,10))
plt.show()
```

```python
'''Ration of Books Published'''

genre1=df['Year'].value_counts()
content=list(genre1.keys())

plt.pie(genre1,labels = content,autopct = '%.2f%%')
plt.title('Percentage Of Books Published', fontsize = 20)
plt.show()
```

```python
'''Scatter Plot'''

sns.scatterplot(x=df.Year, y=df['User Rating'], hue=df.Reviews, s=500);
```

```python
'''Most Expensive Books'''

exp_book = df.sort_values("Price", ascending = False).head(20)

plt.title('Expensive Books')
sns.barplot(y=exp_book.Name, x=exp_book.Price);
```
