```python
#Part 9
#Recommended structure of your code to define logistic regression:
#Getting Errors in these, Need to complete this task.
class LogitRegression:
  iterations = 0
  learning_rate = 0
  def __init__(self,lr=1e-6, iter = 100000):
    self.iterations = iter
    self.learning_rate = lr
    pass
  def fit(self, x, y):
    # This method performs the training.
    # Initialize weights
    # For a number of iterations
    # Call gradient_descent function
    # Call cost function and keep it in an array , e.g.loss.append()
    costs = []
    wt = []
    N = len(y)
    w = np.random.uniform(0,1,7)
    wt.append(w)
    b = 0
    gd = 0
    for i in range(self.iterations):
      w = wt[i]
      t_cosf = 0
      for j in range(N):
        if j >= len(x) or j >= len(y):
          break
        t_cosf += self.cost(x[j], y[j], w, b)
        gd += self.gradient_descent(w, x[j], y[j], b)

      gd = gd/N
      t_cosf = t_cosf/N
      w = w - np.array([self.learning_rate * gd]*7)
      wt.append(w)
      costs.append(t_cosf)
    return wt, costs

  def sigmoid(self, z):
    return 1/ (1 + pow(e,-z))
    
  def cost(self, x ,y, w, b):
    wT = w.transpose()
    h = self.sigmoid(np.matmul(x,wT) + b)
    if h == 1.0:
      return 0
    c =-y * np.log(h) - (1-y) * np.log(1-h)
    return c

  def gradient_descent(self, w, x, y, b):
    wT = w.transpose()
    y_ = self.sigmoid(np.matmul(x,wT) + b)
    xT = x.transpose()
    gd = xT* (y_ - y)
    return gd

  def predict(self, w, x, b):
    y_ = self.sigmoid(np.dot(x,w.transpose()) + b)
    if y_ >= 0.5:
      return 1
    return 0
```
